```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as py
```

## General Funcs

```python
def divergence(total,split):
    
    rand=np.random.random(split)
    
    randN = rand/np.sum(rand) # Normalized
    
    splitT = total*randN
    
    return splitT
    
    # print splitT, maxN
    
def evolution(total,split,mani,generation):
    
    evo = np.asarray([ np.amax(divergence(total,split)) ])
    
    newTotal = np.amax(divergence(total,split)) * mani
    
    for t in np.linspace(1,generation-1,generation-1):
        evo = np.append(evo,np.amax(divergence(newTotal,split)) )
        newTotal = np.amax(divergence(newTotal,split))*mani
        
    return evo
```

```python
divergence(10,5)
```

```python
evolution(10,5,3,100)
```

## Data

Run multiple times and compare data

```python
data = np.asarray([])
for i in np.linspace(1,10,10):
    data =  np.append(data,evolution(10,5,3,100))
```

```python
data = data.reshape( (10,100) )
runN = np.linspace(1,100,100)
runN.shape
```

```python
plt.figure(figsize=(20,12.35))
plt.plot(runN,data[0],'g-', label = "Run # 1")
plt.plot(runN,data[1],'b-', label = "Run # 2")
plt.plot(runN,data[2],'r-', label = "Run # 3")
plt.plot(runN,data[3],'m-', label = "Run # 4")
plt.legend()
plt.show()

# py.iplot_mpl(plt.gcf(),filename="evolutionSim100")
```
