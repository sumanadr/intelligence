## General Funcs

```python
import random

def randomSplit(n, total):  # http://stackoverflow.com/questions/3589214/generate-multiple-random-numbers-to-equal-a-value-in-python
    """Return a randomly chosen list of n positive integers summing to total.
    Each such list is equally likely to occur."""

    dividers = sorted(random.sample(xrange(1, total), n - 1))
    return [a - b for a, b in zip(dividers + [total], [0] + dividers)]
```

```python
test = np.asarray(randomSplit(5,100))*1.0
print test
print test/100
```

```python
mu = 0.5
sigma = 0.35
scale = 1

def divergence(total,split):
    
    # rand=np.random.random(split)
    # rand=np.random.normal(mu,sigma,split)
    # rand=np.random.exponential(scale,split)
    
    # randN = rand/np.sum(rand) # Normalized
    randN = 1.0 * np.asarray(randomSplit(split,100))/100
    
    splitT = total*randN
    
    return splitT
    
    # print splitT, maxN
    
def evolution(total,split,mani,generation):
    
    evo = np.asarray([ np.amax(divergence(total,split)) ])
    
    newTotal = np.amax(divergence(total,split)) * mani
    
    for t in np.linspace(1,generation-1,generation-1):
        evo = np.append(evo,np.amax(divergence(newTotal,split)) )
        newTotal = np.amax(divergence(newTotal,split))*mani
        
    return evo
```

```python
divergence(10,5)
```

```python
evolution(10,5,3,100)
```

## Data

Run multiple times and compare data

```python
data = np.asarray([])
for i in np.linspace(1,10,10):
    data =  np.append(data,evolution(10,5,3,100))  # 10： 初始数量; 5: 分几份; 3:得到最大的之后翻几倍；100：进行100轮 
```

```python
data = data.reshape( (10,100) )
runN = np.linspace(1,100,100)
runN.shape
```

```python
plt.figure(figsize=(20,12.35))
plt.plot(runN,data[0],'g-', label = "Run # 1")
plt.plot(runN,data[1],'b-', label = "Run # 2")
plt.plot(runN,data[2],'r-', label = "Run # 3")
plt.plot(runN,data[3],'m-', label = "Run # 4")
plt.plot(runN,data[4],'g--', label = "Run # 5")
plt.plot(runN,data[5],'b--', label = "Run # 6")

plt.legend()
plt.yscale('log')

plt.show()

# py.iplot_mpl(plt.gcf(),filename="evolutionSim100")
```

```python
# write a new function
def game(init,split,mani,gen,run):
    dataGame = np.asarray([])
    
    for i in np.linspace(1,run,run):
        dataGame=np.append(dataGame,evolution(init,split,mani,gen))
        
    dataGame = dataGame.reshape( (run,gen) )
    # genN = np.linspace(1,gen,gen)
    
    return dataGame
```

```python
game(0.5,5,3,20,5)
```
